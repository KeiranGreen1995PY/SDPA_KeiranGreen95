```python
## Step 1: Crawl a real-world data set (2%) 
## Import pandas data libary and requests to pull the global covid-19 data from the webpage url
import requests
import pandas as pd
```

```python
# Create a variable named url to our target webpage url
# The data set is extracted from an API on covid-19 data
# The variables of interest include new confirmed cases by country , total cases by country, New deaths by country, total deaths by country.
url = "https://api.covid19api.com/summary"
```

```python
# Using the requests package imported , this is to make a GET request from the API endpoint.
# Set the request equal to the variable r 
r = requests.get(url)
```

```python
#Print r to see what has been stoed in r. If the  response code is 200, the request has succeeded
r
```

```python
# Using the json method, extract the  structured data from the request. 
json = r.json()
```

## Question 1
# Is new covid-19 related deaths likely influenced by New confirmed covid-19 cases?


```python

reg1 = 'NewDeaths~NewConfirmed+TotalConfirmed+TotalDeaths' # OLS multiple regression model 
reg1output=smf.ols(reg1,covid_df).fit()
print(reg1output.summary())
```

# Analysis 
Total Confirmed cases, and Total deaths on New Deaths
First of since R-squared value is 96.5% , this means our model is a good fit. 
96.5% of the variation in independant variable New confirmed, total confirmed and new deaths explains the variation in the dependant variable New deaths
Since the p-value of Total confirmed is 0.370, and the p-value of total deaths is 0.206. these two regression coefficients are not statistically significant at the 5% significance level.
Since the p-value of New confirmed cases is 0.000 , it is below the 5% significance level.
New confirmed cases shows a statistically significant impact on New deaths.
