```python
## Step 1: Crawl a real-world data set (2%) 
## Import pandas data libary and requests to pull the global covid-19 data from the webpage url
import requests
import pandas as pd
```

```python
# Create a variable named url to our target webpage url
# The data set is extracted from an API on covid-19 data
# The variables of interest include new confirmed cases by country , total cases by country, New deaths by country, total deaths by country.
url = "https://api.covid19api.com/summary"
```

```python
# Using the requests package imported , this is to make a GET request from the API endpoint.
# Set the request equal to the variable r 
r = requests.get(url)
```

```python
#Print r to see what has been stoed in r. If the  response code is 200, the request has succeeded
r
```

```python
# Using the json method, extract the  structured data from the request. 
json = r.json()
```

```python
# Print the json variable in order to view the structured data from the request. 
json
```

```python
# Using the keys() method to explore the json variable and keys of the dictionary as a list.
json.keys()
```

```python
## Load the dataset in to a covid-19 dataframe in order to prepare the global covid19 data for exploratory data analysis 5%
covid_df = pd.DataFrame(json['Countries'])
```

```python
covid_df # Print the data frame to confirm that the dataset has susscfully been loaded in to the data frame
         # There is no missing data to handle
         # There are no outliers or inconsistencies in the data 
```

```python
# Reorder dataframe by creating a list
# This is to remove ID, Country Code, Slug, Date and Premium columns 
# which are not necessary for the purpose of this analysis.
order = ['Date','Country','NewConfirmed','TotalConfirmed','NewDeaths','TotalDeaths','NewRecovered','TotalRecovered']
covid_df = covid_df[order]
```
